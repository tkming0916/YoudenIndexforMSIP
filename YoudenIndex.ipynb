MSIPの検証結果からROC曲線の描画とYoudenIndexによる閾値の算出

```python
import csv
import pprint
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt

font = {"family": "TakaoGothic"}
plt.rc('font', **font)

# csvの読み込み
data_datadriven = pd.read_csv("./data/hogehoge.csv",encoding="cp932")
data_knowledge = pd.read_csv("./data/hogehoge.csv",encoding="cp932")
```

データ学習型

```python
datadriven = data_datadriven[['#判定　血圧.org', '#判定　脂質.org', '#判定　糖代謝.org', '判定　血圧.異常あり', '判定　脂質.異常あり', '判定　糖代謝.異常あり']]
print(datadriven["#判定　血圧.org"].value_counts())
print(datadriven["#判定　脂質.org"].value_counts())
print(datadriven["#判定　糖代謝.org"].value_counts())

datadriven.loc[datadriven["#判定　血圧.org"]=="異常あり","判定　血圧"]=1
datadriven.loc[datadriven["#判定　血圧.org"]=="異常なし","判定　血圧"]=0

datadriven.loc[datadriven["#判定　脂質.org"]=="異常あり","判定　脂質"]=1
datadriven.loc[datadriven["#判定　脂質.org"]=="異常なし","判定　脂質"]=0

datadriven.loc[datadriven["#判定　糖代謝.org"]=="異常あり","判定　糖代謝"]=1
datadriven.loc[datadriven["#判定　糖代謝.org"]=="異常なし","判定　糖代謝"]=0

print("=== 血圧 ===")
y_score = datadriven["判定　血圧.異常あり"]
y_true = datadriven["判定　血圧"]

fpr, tpr, thresholds = metrics.roc_curve(y_true, y_score)
auc = metrics.auc(fpr, tpr)
print(auc)
Youden_index_candidates = tpr-fpr    
index = np.where(Youden_index_candidates==max(Youden_index_candidates))[0][0]

cutoff = thresholds[index]
kando = tpr[index]
tokuido = 1 - fpr[index]
print(cutoff,"感度:",kando,"特異度：",tokuido)

plt.figure(figsize=(10,10), dpi = 120)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("データ学習型における「高血圧症」のROC曲線",fontsize=20)
plt.plot(fpr, tpr, label='ROC curve (area = %.2f)'%auc)
plt.legend(fontsize=20)
plt.xlabel('偽陽性率(FPR: False positive rate)',fontsize=20)
plt.ylabel('真陽性率(TPR: True positive rate)',fontsize=20)
plt.grid()
# plt.savefig("roc_data_blood.eps")
plt.show()

print("=== 脂質 ===")
y_score = datadriven["判定　脂質.異常あり"]
y_true = datadriven["判定　脂質"]

fpr, tpr, thresholds = metrics.roc_curve(y_true, y_score)
auc = metrics.auc(fpr, tpr)
print(auc)
Youden_index_candidates = tpr-fpr    
index = np.where(Youden_index_candidates==max(Youden_index_candidates))[0][0]

cutoff = thresholds[index]
kando = tpr[index]
tokuido = 1 - fpr[index]
print(cutoff,"感度:",kando,"特異度：",tokuido)

plt.figure(figsize=(10,10), dpi = 120)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("データ学習型における「脂質異常症」のROC曲線",fontsize=20)
plt.plot(fpr, tpr, label='ROC curve (area = %.2f)'%auc)
plt.legend(fontsize=20)
plt.xlabel('偽陽性率(FPR: False positive rate)',fontsize=20)
plt.ylabel('真陽性率(TPR: True positive rate)',fontsize=20)
plt.grid()
# plt.savefig("roc_data_dyslipidemia.eps")
plt.show()

print("=== 糖代謝 ===")
y_score = datadriven["判定　糖代謝.異常あり"]
y_true = datadriven["判定　糖代謝"]

fpr, tpr, thresholds = metrics.roc_curve(y_true, y_score)
auc = metrics.auc(fpr, tpr)
print(auc)
Youden_index_candidates = tpr-fpr    
index = np.where(Youden_index_candidates==max(Youden_index_candidates))[0][0]

cutoff = thresholds[index]
kando = tpr[index]
tokuido = 1 - fpr[index]
print(cutoff,"感度:",kando,"特異度：",tokuido)

plt.figure(figsize=(10,10), dpi = 120)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("データ学習型における「糖代謝」のROC曲線",fontsize=20)
plt.plot(fpr, tpr, label='ROC curve (area = %.2f)'%auc)
plt.legend(fontsize=20)
plt.xlabel('偽陽性率(FPR: False positive rate)',fontsize=20)
plt.ylabel('真陽性率(TPR: True positive rate)',fontsize=20)
plt.grid()
# plt.savefig("roc_data_sugarmetabolism.eps")
plt.show()
```
